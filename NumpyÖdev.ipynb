```python
#21) Bu matrisin her bir sütununa, o sütunun ortalama değerini çıkararak veriyi merkezleyin (ortalama sıfır olacak şekilde).
arr_21=np.random.randint(1,26,(5,5))
print( arr_21)

a=arr_21.mean(axis=0)
print(a)

arr_21_centered = arr_21 - a
print( arr_21_centered)

print( arr_21_centered.mean(axis=0))
```

```python
#22)	C = np.array([[4, 7], [2, 6]]) matrisini oluşturun.
C = np.array([[4, 7], [2, 6]])
```

```python
#23)	Matrisin determinantını hesaplayın.
print(np.linalg.det(C))
```

```python
#24)	Matrisin tersini (inverse) hesaplayın ve ters matris ile orijinal matrisi çarparak birim matris (Identity Matrix) elde edip etmediğinizi kontrol edin.
Ters=np.linalg.inv(C)
print(Ters)

print(Ters*C)
```

```python
#28)	v1 = np.array([1, 2, 3]) ve v2 = np.array([4, 5, 6]) vektörlerini oluşturun.
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
```

```python
#29)	Bu iki vektörün nokta ürününü (inner product) ve dış ürününü (outer product) hesaplayın.

inner_product = np.dot(v1, v2)
print("Inner Product (Nokta Ürünü):", inner_product)

outer_product = np.outer(v1, v2)
print("\nOuter Product (Dış Ürünü):\n", outer_product)
```

```python
#30)	1.000.000 elemanlı büyük bir listeyi (py_list) ve aynı elemanları içeren bir NumPy dizisini (np_array) oluşturun
```

```python
#30) 1.000.000 elemanlı büyük bir listeyi (py_list) ve aynı elemanları içeren bir NumPy dizisini (np_array) oluşturun

import time

start_time = time.time()
py_list = list(range(1000000))
end_time = time.time()
print(f"Python list created in {end_time - start_time:.4f} seconds.")

start_time = time.time()
np_array = np.arange(1000000)
end_time = time.time()
print(f"NumPy array created in {end_time - start_time:.4f} seconds.")

# Display the first few elements and the size to confirm
print("\nFirst 5 elements of py_list:", py_list[:5])
print("Size of py_list:", len(py_list))
print("\nFirst 5 elements of np_array:", np_array[:5])
print("Size of np_array:", np_array.size)
```
